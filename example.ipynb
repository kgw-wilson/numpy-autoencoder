The following experiments will show off the
autoencoder using both relu and sigmoid 
activation functions. Visualizations are
from the first 5 images in the test set.

```python
from autoencoder import AutoEncoder
```

```python
# Constants for experiments with both activation functions
INPUT_DIM = 28**2
BATCH_SIZE = 64
```

```python
# Model with ReLU Activation function

model = AutoEncoder(
    epochs=50,
    batch_size=BATCH_SIZE,
    learning_rate=0.005,
    encoder_layer_sizes=[INPUT_DIM, 256, 64],
    decoder_layer_sizes=[64, 256, INPUT_DIM],
    latent_layer_size=16,
    activation_func="relu",
)

losses = model.train()
model.show_loss_curve(losses)
model.visualize()
```

```python
# Model with Sigmoid Activation function

model = AutoEncoder(
    epochs=50,
    batch_size=BATCH_SIZE,
    learning_rate=0.05,
    encoder_layer_sizes=[INPUT_DIM, 128, 32],
    decoder_layer_sizes=[32, 128, INPUT_DIM],
    latent_layer_size=16,
    activation_func="sigmoid",
)

losses = model.train()
model.show_loss_curve(losses)
model.visualize()
```
